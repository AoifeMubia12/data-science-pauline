#### *Chaining Operations*
##### 1.Filter for respondents over 40 years old
##### 2.Select only columns for degree, sex, income
##### 3.Create a new column "income_category" that labels income as 'High' if over $50000, otherwise 'low'
##### 4.Sort results by income in descending order(highest incomes first)

```python
gss_processed = (
    gss.query("age > 40")
    .loc[:, ['degree', 'income', 'sex']]
    .assign(income_category = lambda x: np.where(x['income'] > 50000, 'High', 'Low'))
    .sort_values(by = 'income', ascending = False))
gss_processed
```

```python
gss_processed = (
    gss.query("age > 40")
    .loc[:, ['degree', 'income', 'age']]
    .assign(age_category = lambda x: np.where(x['age'] > 25, 'Adults', 'Minor'))
    .sort_values(by = 'income', ascending = False))
gss_processed
```
